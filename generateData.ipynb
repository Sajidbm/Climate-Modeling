Here we will implement a data generation pipeline in JAX. We need a few things components:
- noise.py : generates Gaussian noise for process and observation models

For now, we will not introduce any process noise into our dynamics. 

```python
import jax
import jax.numpy as jnp
from noise import GaussianNoise
from Study.lorenz63JAX import rollout, LorenzSimulatorK4

# lorenz63 system
u_0_set = jax.random.normal(jax.random.PRNGKey(0), (9, 3))
lorenzStepper = LorenzSimulatorK4()
iterations = 7000

rollout_func = rollout(lorenzStepper, iterations, include_init=False)
dataset = jax.vmap(rollout_func)(u_0_set)

# observation and process noise
obs_noise = GaussianNoise(sigma = 1.0)
process_noise = None

# create noisy dataset
key = jax.random.PRNGKey(42)
noisy_dataset = obs_noise(key, dataset)
```
